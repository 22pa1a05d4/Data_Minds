```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark=SparkSession.builder.getOrCreate()
```

```python
df_encounters=spark.read.csv('/Volumes/data_minds/bronze/raw_data/encounters.csv',header=True,inferSchema=True)
df_organizations=spark.read.csv('/Volumes/data_minds/bronze/raw_data/organizations.csv',header=True,inferSchema=True)
df_patients=spark.read.csv('/Volumes/data_minds/bronze/raw_data/patients.csv',header=True,inferSchema=True)
df_payers=spark.read.csv('/Volumes/data_minds/bronze/raw_data/payers.csv',header=True,inferSchema=True)
df_procedures=spark.read.csv('/Volumes/data_minds/bronze/raw_data/procedures.csv',header=True,inferSchema=True)
```

```python
df_encounters.write.format('delta').mode('overwrite').save("/Volumes/data_minds/bronze/raw_data/Bronze_encounters")
df_organizations.write.format('delta').mode('overwrite').save("/Volumes/data_minds/bronze/raw_data/Bronze_organizations")
df_patients.write.format('delta').mode('overwrite').save("/Volumes/data_minds/bronze/raw_data/Bronze_patients")
df_payers.write.format('delta').mode('overwrite').save("/Volumes/data_minds/bronze/raw_data/Bronze_payers")
df_procedures.write.format('delta').mode('overwrite').save("/Volumes/data_minds/bronze/raw_data/Bronze_procedures")
```
